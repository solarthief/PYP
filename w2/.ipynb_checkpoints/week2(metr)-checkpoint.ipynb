```python
from pandas import DataFrame, Series 
import pandas as pd
import numpy as np
import matplotlib
import sklearn.model_selection
import sklearn.neighbors
import sklearn.preprocessing
import sklearn.datasets
```

```python
dat=sklearn.datasets.load_boston()
```

```python
attributes=sklearn.preprocessing.scale(dat.data)
classes=dat.target
results=DataFrame()
```

```python
kf=sklearn.model_selection.KFold(n_splits=5,random_state=42,shuffle=True)
for i in np.linspace(1,10,200):
    neigh=sklearn.neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance',p=i)
    cr_val=sklearn.model_selection.cross_val_score(neigh,attributes,classes,cv=kf,scoring='neg_mean_squared_error')
    results=results.append([(i,cr_val.mean())],ignore_index=True)
```
